# Machine Learning

- Extracting patterns from a data is called Machine Learning.
- Data is of two types:
  1. Features - Inputs - All the Information about the Object.
  2. Target - To Predict - The variable we want to predict about the object.
- Output: **Model** - It encapsulates all the patterns for the specific problem. Model can be used later on different datasets for the same problems.

#### Rules-based Systems vs. Machine Learning:

- Rules-based System: Data -> Extract the Rules -> Code the Software -> Outcomes.
- Rules keeps changing. Codes keeps building. A never ending cumbersome process.
- In Machine Learning, Data + Model = Outcome.
- Rules in a Rules-based System becomes the features for a Machine Learning Model.
- We use Probabilities to finally decide upon the machine learning outputs if they align the correct values or not.

## Supervised Machine Learning

The Goal of ***Supervised Machine Learning*** is to come up with a function ***g( )*** that takes in the Feature Matrix **X**, and produces something that is as close as possible to the target variable **y**.

***g*( X ) ~= y**
- **X** -> Feature Matrix
- ***g( )*** -> Model Engine / Mathematical Function
- **y** -> Target Vaiable / Variable to Predict

The process of looking up at features and coming-up with a model function ***g( )*** is called **Training**.

#### Types of Supervised Machine Learning
Supervised Machine Learning problems are categorised into the following, based on the kind of output desired of the model, ***g( )***:

*   Regression -> Anything that outputs a number from - infinity to + infinity, it is a Regression problem.
    - Linear
    - Non-Linear
    - Logistic
*   Classification -> Here, the output of the function is a category.
    - Binary Classification -> Output is binary, i.e. a 0 or a 1.
    - Multi-Class Classification -> Output can be one among the many definite classes, e.g. cats, rats, dogs, bats, etc.
*   Ranking & Scoring -> Outputs the rankings of objects, categories, classes, or data as input. Used in building recommendation systems, scoring website search results by Google, e-Commerce, etc.

Modelling:
  * Actual Machine Learning happens here!
  * Try different models (e.g. Linear Regression, Decision Trees, Random Forests, Neural Networks, etc.)
    * Select different machine learning models and run each one of them on the data
    * Redo Preprocessing if faced with any data issues, or discover any discrepancies
  * Select the best one -> the best one makes our ***g( )***

Modelling: Model Selection Process:
  * Split the available data into three non-overlaping parts.
  * These parts are: Training Dataset, Validation Dataset, and Test Dataset.
  * The splitting may be done in the ratios of 60:20:20, 80:10:10, 90:5:5 or others.
  * Train all the shortlisted models on the training dataset.
  * Run all the models on the validation dataset. Pick the one with the best accuracy value/percentage.
  * Run the selected model on the test dataset and check if the obtained accuracy percentage is close to the value obtained on the validation dataset.
  * If yes, then select this model. Otherwise, look for another model.

Modelling: Final
  * Train the final selected model on the Training + Validation combined dataset.
  * Run the newly trained model on the test dataset and evaluate the outcomes.
  * Use this model for final deployment.
